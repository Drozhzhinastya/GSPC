```python
with open('triod_tsvetnaya_full.txt', 'r', encoding='utf-8') as f:
    triod = f.readlines()
```

```python
triod[:10]
```

```python
import re
import pandas as pd
```

```python
df = pd.DataFrame(columns=['red_text_b', 'black_text', 'red_text_e', 'color'])
```

```python
df
```

```python
i = 0
counter_black = 0

black_elem = []

for line in triod:
    try:
        if line.startswith('<'):
            df = df.append({"red_text_b":line, "black_text":'', 'red_text_e':'', "color": 'red'}, ignore_index=True)
            if counter_black >= 1:
                string = ''.join(black_elem)
                df = df.append({"red_text_b":'', "black_text":string, 'red_text_e':'', "color": 'black'}, ignore_index=True)
                black_elem = []
                counter_black = 0
        
        elif '<' in line:
            #киноварь внутри строки - отделяем служебные пометы
            k_raw_text = re.search(r'(?<=<).+(?=>)', line)
            k_text = k_raw_text.group(0)
            k_clean_text = re.sub(r'[\(\)\[\]]', '', k_text) #служебная помета

            black_text = re.sub(r'(?<=<).+(?=>)', '', line)
            black_text = re.sub(r'[\(\)\[\]\<\>\']', '', black_text)
            
            df = df.append({"red_text_b":'', "black_text":black_text, 'red_text_e':'', "color": 'black'}, ignore_index=True)
            df = df.append({"red_text_b":'', "black_text":'', 'red_text_e':k_clean_text,"color": 'red'}, ignore_index=True)

        else:
            if '#' not in line:
                black_elem.append(line)
                counter_black+=1
            else:
                df = df.append({"red_text_b":line, "black_text":'', 'red_text_e':'', "color": 'temp'}, ignore_index=True)
    except:
        print(line)
            
    i+=1

print(i)
```
